```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv('../input/topreee/Titanic.csv')
df
```

```python
df1=pd.get_dummies(df)
df1
```

```python
df2 = apriori(df1, min_support = 0.3, use_colnames = True)
df2
```

```python
df_ar = association_rules(df2, metric = "confidence", min_threshold = 0.6)
df_ar
```

```python
df_ar[ (df_ar['lift'] >= 1) &
       (df_ar['confidence'] >= 0.8) ]
```
